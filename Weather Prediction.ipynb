# Problem Statement - Given Max Temperature as an independent feature we have to predict Min Temperature(target feature or dependent variable)

```python
#Independent feature
X = df1['MaxTemp'].values.reshape(-1,1)     
#Since we are considering only one feature we had to apply reshape to make it an array

# Dependent feature
y = df1['MinTemp'].values.reshape(-1,1)
```

```python
#Train_test split 70:30
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Object class
lr = LinearRegression()
```

```python
#Train
lr.fit(X_train,y_train)
```

```python
lr_pred = lr.predict(X_test)   #Predictions of mintemp
```

```python
lr_pred
```

```python
from sklearn.metrics import r2_score
```

```python
#Checking accuracy score
r2_score(y_test,lr_pred)*100
```

Means our accuracy is 76.5%. To cross check whether r2 score is correct,the r2score must be same as the accuracy score of test data

```python
#Accuracy calculation of test data
lr.score(X_test,y_test)*100
```

```python
#Accuracy calculation of train data
lr.score(X_train,y_train)*100
```

The model is not overfitted because train and  test accuracy are besides
